```python
import pandas as pd
import datetime
import numpy as np
```

```python
df = pd.read_csv('^NSEI.csv', parse_dates=True)
```

```python
df
```

```python
df['sma50'] = df['Adj Close'].rolling(50).mean()
df['sma200'] = df['Adj Close'].rolling(200).mean()
```

```python
df = df.dropna()
```

```python
df
```

```python
df['status'] = np.where(df['sma50'] > df['sma200'], 'buy_zone','sell_zone')
```

```python
df
```

```python
position = 0

trades = []

for i in range(len(df)):
    
    if ((df['sma50'].iloc[i] > df['sma200'].iloc[i]) and (position != 1)):
        position = 1
        trades.append((df['Adj Close'].iloc[i], 'BUY'))
    elif ((df['sma200'].iloc[i] > df['sma50'].iloc[i]) and (position != -1)):
        position = -1
        trades.append((df['Adj Close'].iloc[i], 'SELL'))
    else:
        pass


```

```python
def backtest(p1, p2, df):
    df[f'sma{p1}'] = df['Adj Close'].rolling(p1).mean()
    df[f'sma{p2}'] = df['Adj Close'].rolling(p2).mean()

    df = df.dropna()

    position = 0

    trades = []

    for i in range(len(df)):

        if ((df[f'sma{p1}'].iloc[i] > df[f'sma{p2}'].iloc[i]) and (position != 1)):
            position = 1
            trades.append((df['Adj Close'].iloc[i], 'BUY'))
        elif ((df[f'sma{p2}'].iloc[i] > df[f'sma{p1}'].iloc[i]) and (position != -1)):
            position = -1
            trades.append((df['Adj Close'].iloc[i], 'SELL'))
        else:
            pass

    return trades

```

```python
backtest(40, 200, df)
```
