

### Z Values

Standard scores or z-scores, are a statistical measure that quantifies how many standard deviations an observation or data point is away from the mean of a distribution


Z-Test-Statistics-Formula.jpg

```python
# Let's Understand using an example

# Let there be two students in two separate classes

class_gaurav_scores = np.random.normal(loc=80, scale=10, size=100).astype(int)    # Mean = 80, Standard Deviation = 10
class_saurabh_scores = np.random.normal(loc=75, scale=8, size=100).astype(int)       # Mean = 75, Standard Deviation = 8
```

```python
gaurav_score = 82
saurabh_score = 78
```

```python
# calculate z scores for both the students

gaurav_z_scores = (gaurav_score - 80) / (10)
saurabh_z_scores = (saurabh_score - 75) / (8)
```

```python
gaurav_z_scores, saurabh_z_scores
```

**Lets see the competition Visually**

```python
# Calculate the corresponding value on the normal distribution PDF
pdf_value_gaurav = stats.norm.pdf(gaurav_z_scores)
pdf_value_saurabh = stats.norm.pdf(saurabh_z_scores)

# Generate x-values for the normal distribution curve
x = np.linspace(-4, 4, 100)
y = stats.norm.pdf(x)

# Plot the normal distribution curve
plt.plot(x, y, label='Normal Distribution')

# Highlight the z-value with a vertical line
plt.axvline(x=gaurav_z_scores, color='red', linestyle='--', label='z = 0.2')

plt.axvline(x=saurabh_z_scores, color='blue', linestyle='--', label='z = 0.375')

# Add labels and legend
plt.xlabel('z score compare')
plt.ylabel('Probability Density')
plt.legend()

# Display the plot
plt.show()
```

**Hypothesis Testing**


Hypothesis testing is a statistical method used to make inferences and draw conclusions about a population based on a sample of data. It involves formulating two competing hypotheses, known as the null hypothesis (H0) and the alternative hypothesis (H1 or Ha), and performing statistical tests to evaluate the evidence against the null hypothesis.


Some Key Terms

1. Null Hypothesis
2. Alternate Hypothesis
3. One Tailed Test
4. Two Tailed Test
5. Significance Level
6. p-value
7. Type I error
8. Type II error
9. Critical Value
10. Power
